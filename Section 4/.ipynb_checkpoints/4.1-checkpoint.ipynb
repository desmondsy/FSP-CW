```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import matplotlib.pyplot as pylab
import seaborn as sns
from datetime import datetime
```

```python
params = {'figure.figsize': (15, 5),
         'axes.labelsize': 14,
         'axes.titlesize': 14,
          'axes.titlesize': 18}
pylab.rcParams.update(params)
```

```python
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
```

# 4.1.1

```python
mean_aapl, std_aapl, median_aapl = np.mean(aapl),np.std(aapl),np.median(aapl)
mean_ibm, std_ibm, median_ibm = np.mean(ibm),np.std(ibm),np.median(ibm)
mean_jpm, std_jpm, median_jpm = np.mean(jpm),np.std(jpm),np.median(jpm)
mean_djx, std_djx, median_djx = np.mean(djx),np.std(djx),np.median(djx)
```

```python
aapl['Adj Returns'] = aapl['Adj Close'].pct_change()*100
ibm['Adj Returns'] = ibm['Adj Close'].pct_change()*100
jpm['Adj Returns'] = jpm['Adj Close'].pct_change()*100
djx['Adj Returns'] = djx['Adj Close'].pct_change()*100
aapl
```

# 4.1.3

```python
#rolling means
w = 5
aapl_rolling_mean = aapl['Adj Close'].rolling(w).mean().dropna()
ibm_rolling_mean = ibm['Adj Close'].rolling(w).mean().dropna()
jpm_rolling_mean = jpm['Adj Close'].rolling(w).mean().dropna()
djx_rolling_mean = djx['Adj Close'].rolling(w).mean().dropna()
#rolling median
aapl_rolling_med = aapl['Adj Close'].rolling(w).median().dropna()
ibm_rolling_med = ibm['Adj Close'].rolling(w).median().dropna()
jpm_rolling_med = jpm['Adj Close'].rolling(w).median().dropna()
djx_rolling_med = djx['Adj Close'].rolling(w).median().dropna()
#mean absolute deviation
aapl_mad = (abs(aapl_rolling_med - aapl_rolling_med.rolling(5).median())).median()
ibm_mad = (abs(ibm_rolling_med - ibm_rolling_med.rolling(5).median())).median()
jpm_mad = (abs(jpm_rolling_med - jpm_rolling_med.rolling(5).median())).median()
djx_mad = (abs(djx_rolling_med - djx_rolling_med.rolling(5).median())).median()
aapl_mad
```

```python
#AAPL
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(aapl['Adj Close'])
plt.plot(aapl_rolling_mean)
plt.plot(aapl_rolling_mean+1.5*aapl_rolling_mean.rolling(5).std())
plt.plot(aapl_rolling_mean-1.5*aapl_rolling_mean.rolling(5).std())
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling mean','Upper bound (SD)','Lower bound (SD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('AAPL rolling mean with +/- 1.5x SD bounds')
plt.subplot(122)
plt.plot(aapl['Adj Close'])
plt.plot(aapl_rolling_med)
plt.plot(aapl_rolling_med - 1.5*aapl_mad)
plt.plot(aapl_rolling_med + 1.5*aapl_mad)
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling median','Upper bound (MAD)','Lower bound (MAD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('AAPL rolling median with +/- 1.5x MAD bounds')
#IBM
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(ibm['Adj Close'])
plt.plot(ibm_rolling_mean)
plt.plot(ibm_rolling_mean+1.5*ibm_rolling_mean.rolling(5).std())
plt.plot(ibm_rolling_mean-1.5*ibm_rolling_mean.rolling(5).std())
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling mean','Upper bound (SD)','Lower bound (SD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('IBM rolling mean with +/- 1.5x SD bounds')
plt.subplot(122)
plt.plot(ibm['Adj Close'])
plt.plot(ibm_rolling_med)
plt.plot(ibm_rolling_med - 1.5*ibm_mad)
plt.plot(ibm_rolling_med + 1.5*ibm_mad)
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling median','Upper bound (MAD)','Lower bound (MAD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('IBM rolling median with +/- 1.5x MAD bounds')
#JPM
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(jpm['Adj Close'])
plt.plot(jpm_rolling_mean)
plt.plot(jpm_rolling_mean+1.5*jpm_rolling_mean.rolling(5).std())
plt.plot(jpm_rolling_mean-1.5*jpm_rolling_mean.rolling(5).std())
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling mean','Upper bound (SD)','Lower bound (SD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('JPM rolling mean with +/- 1.5x SD bounds')
plt.subplot(122)
plt.plot(jpm['Adj Close'])
plt.plot(jpm_rolling_med)
plt.plot(jpm_rolling_med - 1.5*jpm_mad)
plt.plot(jpm_rolling_med + 1.5*jpm_mad)
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling median','Upper bound (MAD)','Lower bound (MAD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('JPM rolling median with +/- 1.5x MAD bounds')
#DJX
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(djx['Adj Close'])
plt.plot(djx_rolling_mean)
plt.plot(djx_rolling_mean+1.5*djx_rolling_mean.rolling(5).std())
plt.plot(djx_rolling_mean-1.5*djx_rolling_mean.rolling(5).std())
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling mean','Upper bound (SD)','Lower bound (SD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('DJX rolling mean with +/- 1.5x SD bounds')
plt.subplot(122)
plt.plot(djx['Adj Close'])
plt.plot(djx_rolling_med)
plt.plot(djx_rolling_med - 1.5*djx_mad)
plt.plot(djx_rolling_med + 1.5*djx_mad)
plt.grid('major',axis='both')
plt.legend(['Adj Close','Rolling median','Upper bound (MAD)','Lower bound (MAD)'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('DJX rolling median with +/- 1.5x MAD bounds')
```

```python
#reset
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
```

```python
outliers = ['2018-05-14','2018-09-14','2018-12-14','2019-01-14']
aapl_max = aapl['Adj Close'].max()*1.2
ibm_max = ibm['Adj Close'].max()*1.2
jpm_max = jpm['Adj Close'].max()*1.2
djx_max = djx['Adj Close'].max()*1.2

for i in outliers:
    aapl.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = aapl_max
    ibm.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = ibm_max
    jpm.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = jpm_max
    djx.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = djx_max
```

```python
#rolling means
w = 5
aapl_rolling_mean = aapl['Adj Close'].rolling(w).mean().dropna()
ibm_rolling_mean = ibm['Adj Close'].rolling(w).mean().dropna()
jpm_rolling_mean = jpm['Adj Close'].rolling(w).mean().dropna()
djx_rolling_mean = djx['Adj Close'].rolling(w).mean().dropna()
#rolling median
aapl_rolling_med = aapl['Adj Close'].rolling(w).median().dropna()
ibm_rolling_med = ibm['Adj Close'].rolling(w).median().dropna()
jpm_rolling_med = jpm['Adj Close'].rolling(w).median().dropna()
djx_rolling_med = djx['Adj Close'].rolling(w).median().dropna()
#mean absolute deviation
aapl_mad = (abs(aapl_rolling_med - aapl_rolling_med.rolling(5).median())).median()
ibm_mad = (abs(ibm_rolling_med - ibm_rolling_med.rolling(5).median())).median()
jpm_mad = (abs(jpm_rolling_med - jpm_rolling_med.rolling(5).median())).median()
djx_mad = (abs(djx_rolling_med - djx_rolling_med.rolling(5).median())).median()
aapl_mad
```
