```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import matplotlib.pyplot as pylab
import seaborn as sns
from datetime import datetime
```

```python
params = {'figure.figsize': (15, 5),
         'axes.labelsize': 14,
         'axes.titlesize': 14,
          'axes.titlesize': 18}
pylab.rcParams.update(params)
```

```python
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
# month = [aapl.index[i].strftime("%Y-%m-%d") for i in np.linspace(0,len(aapl.index)-1,12).astype(int)]
```

```python
plt.figure()
plt.subplot(221)
sns.boxplot(aapl['Adj Close'])
plt.title('AAPL adjusted close boxplot')
plt.subplot(222)
sns.boxplot(ibm['Adj Close'])
plt.title('IBM adjusted close boxplot')
plt.subplot(223)
sns.boxplot(jpm['Adj Close'])
plt.title('JPM adjusted close boxplot')
plt.subplot(224)
sns.boxplot(djx['Adj Close'])
plt.title('DJX adjusted close boxplot')
plt.tight_layout()
#plt.savefig('/Users/desmond/Desktop/FSP_CW/all/Section 4/eps_files/boxplot.eps')
```
