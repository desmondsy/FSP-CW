```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import matplotlib.pyplot as pylab
import random
from datetime import datetime
import seaborn as sns
from sklearn.linear_model import LinearRegression, HuberRegressor
```

```python
params = {'figure.figsize': (20, 5),
         'axes.labelsize': 14,
         'axes.titlesize': 14,
          'axes.titlesize': 18}
pylab.rcParams.update(params)
```

```python
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
```

```python
#Log returns
aapl['Adj Returns'] = np.log(aapl['Adj Close']).diff()
ibm['Adj Returns'] = np.log(ibm['Adj Close']).diff()
jpm['Adj Returns'] = np.log(jpm['Adj Close']).diff()
djx['Adj Returns'] = np.log(djx['Adj Close']).diff()
aapl['Adj Returns'].dropna(inplace=True)
ibm['Adj Returns'].dropna(inplace=True)
jpm['Adj Returns'].dropna(inplace=True)
djx['Adj Returns'].dropna(inplace=True)
```

# no outliers, ols/huber

```python
y = djx['Adj Returns']
x_axis = np.linspace(-0.1,0.1,len(ibm['Adj Returns'])).reshape(-1,1)

#AAPL
OLS,huber = LinearRegression(),HuberRegressor()
OLS.fit(np.array(aapl['Adj Returns']).reshape(-1,1),y)
huber.fit(np.array(aapl['Adj Returns']).reshape(-1,1),y)
plt.figure()
plt.subplot(131)
plt.scatter(aapl['Adj Returns'],y)
plt.plot(x_axis,x_axis*OLS.coef_[0]+OLS.intercept_,'red')
plt.plot(x_axis,x_axis*huber.coef_[0]+huber.intercept_,'green')
plt.grid('major',axis='both')
plt.xlabel('AAPL daily returns')
plt.ylabel('DJIA daily returns')
plt.title('AAPL vs DJIA returns, no outliers')
plt.legend(['OLS','Huber'])
print('AAPL OLS R^2 score: ',OLS.score(np.array(aapl['Adj Returns']).reshape(-1,1),y))
print('AAPL Huber R^2 score: ',huber.score(np.array(aapl['Adj Returns']).reshape(-1,1),y))

#IBM
OLS,huber = LinearRegression(),HuberRegressor()
OLS.fit(np.array(ibm['Adj Returns']).reshape(-1,1),y)
huber.fit(np.array(ibm['Adj Returns']).reshape(-1,1),y)
plt.subplot(132)
plt.scatter(ibm['Adj Returns'],y)
plt.plot(x_axis,x_axis*OLS.coef_[0]+OLS.intercept_,'red')
plt.plot(x_axis,x_axis*huber.coef_[0]+huber.intercept_,'green')
plt.grid('major',axis='both')
plt.xlabel('IBM daily returns')
plt.ylabel('DJIA daily returns')
plt.title('IBM vs DJIA returns, no outliers')
plt.legend(['OLS','Huber'])
print('IBM OLS R^2 score: ',OLS.score(np.array(ibm['Adj Returns']).reshape(-1,1),y))
print('IBM Huber R^2 score: ',huber.score(np.array(ibm['Adj Returns']).reshape(-1,1),y))


#JPM
OLS,huber = LinearRegression(),HuberRegressor()
OLS.fit(np.array(jpm['Adj Returns']).reshape(-1,1),y)
huber.fit(np.array(jpm['Adj Returns']).reshape(-1,1),y)
plt.subplot(133)
plt.scatter(jpm['Adj Returns'],y)
plt.plot(x_axis,x_axis*OLS.coef_[0]+OLS.intercept_,'red')
plt.plot(x_axis,x_axis*huber.coef_[0]+huber.intercept_,'green')
plt.grid('major',axis='both')
plt.xlabel('JPM daily returns')
plt.ylabel('DJIA daily returns')
plt.title('JPM vs DJIA returns, no outliers')
plt.legend(['OLS','Huber'])
print('JPM OLS R^2 score: ',OLS.score(np.array(jpm['Adj Returns']).reshape(-1,1),y))
print('JPM Huber R^2 score: ',huber.score(np.array(jpm['Adj Returns']).reshape(-1,1),y))
#plt.savefig('/Users/desmond/Desktop/FSP_CW/all/Section 4/eps_files/no_outliers.eps')
```

# with outliers 

```python
#reset
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')

outliers = ['2018-05-14','2018-09-14','2018-12-14','2019-01-14']
aapl_max = aapl['Adj Close'].max()*1.2
ibm_max = ibm['Adj Close'].max()*1.2
jpm_max = jpm['Adj Close'].max()*1.2
djx_max = djx['Adj Close'].max()*1.2

for i in outliers:
    aapl.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = aapl_max
    ibm.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = ibm_max
    jpm.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = jpm_max
    djx.loc[datetime.strptime(i,'%Y-%m-%d'),'Adj Close'] = djx_max

aapl['Adj Returns'] = np.log(aapl['Adj Close']).diff()
ibm['Adj Returns'] = np.log(ibm['Adj Close']).diff()
jpm['Adj Returns'] = np.log(jpm['Adj Close']).diff()
djx['Adj Returns'] = np.log(djx['Adj Close']).diff()
aapl['Adj Returns'].dropna(inplace=True)
ibm['Adj Returns'].dropna(inplace=True)
jpm['Adj Returns'].dropna(inplace=True)
djx['Adj Returns'].dropna(inplace=True)
```

```python
#reset
start = '2018-03-16'
end = '2019-03-11'

aapl = data.DataReader("AAPL", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')
ibm = data.DataReader("IBM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
jpm = data.DataReader("JPM", 
                       start=start, 
                       end=end,
                       data_source='yahoo')
djx = data.DataReader("DJIA", 
                       start=start, 
                       end=end, 
                       data_source='yahoo')

aapl_max = aapl['Adj Close'].max()*1.2
ibm_max = ibm['Adj Close'].max()*1.2
jpm_max = jpm['Adj Close'].max()*1.2
djx_max = djx['Adj Close'].max()*1.2

for i in outliers_reg:
    aapl.loc[i,'Adj Close'] = aapl_max
    ibm.loc[i,'Adj Close'] = ibm_max
    jpm.loc[i,'Adj Close'] = jpm_max
    djx.loc[i,'Adj Close'] = djx_max

aapl['Adj Returns'] = np.log(aapl['Adj Close']).diff()
ibm['Adj Returns'] = np.log(ibm['Adj Close']).diff()
jpm['Adj Returns'] = np.log(jpm['Adj Close']).diff()
djx['Adj Returns'] = np.log(djx['Adj Close']).diff()
aapl['Adj Returns'].dropna(inplace=True)
ibm['Adj Returns'].dropna(inplace=True)
jpm['Adj Returns'].dropna(inplace=True)
djx['Adj Returns'].dropna(inplace=True)
```
